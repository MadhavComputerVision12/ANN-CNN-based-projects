```python
import numpy as np

# Define the sigmoid activation function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Create a dataset with input and target values
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Initialize neural network parameters
input_size = 2
hidden_size = 4
output_size = 1
learning_rate = 0.1

# Initialize weights and biases
input_layer = X
weights_input_hidden = np.random.uniform(size=(input_size, hidden_size))
weights_hidden_output = np.random.uniform(size=(hidden_size, output_size))
biases_hidden = np.random.uniform(size=(1, hidden_size))
biases_output = np.random.uniform(size=(1, output_size))

# Training loop
for i in range(10000):
    # Forward propagation
    hidden_layer_input = np.dot(input_layer, weights_input_hidden) + biases_hidden
    hidden_layer_output = sigmoid(hidden_layer_input)
    output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + biases_output
    output_layer_output = sigmoid(output_layer_input)

    # Calculate error using Mean Squared Error
    error = y - output_layer_output
    mse = np.mean(error**2)

    # Backpropagation
    d_output = error * sigmoid_derivative(output_layer_output)
    error_hidden = d_output.dot(weights_hidden_output.T)
    d_hidden = error_hidden * sigmoid_derivative(hidden_layer_output)

    # Update weights and biases
    weights_hidden_output += hidden_layer_output.T.dot(d_output) * learning_rate
    weights_input_hidden += input_layer.T.dot(d_hidden) * learning_rate
    biases_output += np.sum(d_output, axis=0, keepdims=True) * learning_rate
    biases_hidden += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate

print("Final predictions after training:")
print(output_layer_output)
```
